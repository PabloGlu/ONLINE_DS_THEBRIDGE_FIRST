## Ejercicio 1

1. Crea dos variables numericas: un `int` y un `float`
2. Comprueba sus tipos
3. Sumalas en otra nueva
4. ¿De qué tipo es la nueva variable?
5. Elimina las dos primeras variables creadas

```python
var_int = 3
var_float = 5.1

print(f"La variable {var_int} es de clase {type(var_int)} y la variable {var_float} es de tipo {type(var_float)}")

var_sum = var_int + var_float
print(f"La suma {var_sum} es de tipo {type(var_sum)}")

del(var_int, var_float)

try:
   print(var_int)
except:
  print('Se ha eliminado la variable var_int')
try:
   print(var_float)
except:
  print('Se ha eliminado la variable var_float')
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
var_base_in = input("Introduce la base en cm: ")
var_altura_in = input("Introduce la altura en cm: ")

var_base_clean = var_base_in.replace("cm", "").replace(" ", "").replace(",", "").replace("centimetros", "").replace("centímetros", "")
var_altura_clean = var_altura_in.replace("cm", "").replace(" ", "").replace(",", "").replace("centímetros", "")

var_base = float(var_base_clean)
var_altura = float(var_altura_clean)

print(var_base * var_altura)

var_area = var_base * var_altura
print(f"El area es {var_area}cm2")
```
