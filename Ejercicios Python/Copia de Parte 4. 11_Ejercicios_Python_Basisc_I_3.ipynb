## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
vin_1 = input("Primer input: ")
vin_2 = input("Segundo input: ")

vin_x = (vin_1 == vin_2)

if vin_x == True:
  print("¡Tus 2 inputs son... Iguales!")
elif vin_x == False:
  print("¡Tus 2 inputs... No son iguales!")
else:
  print("ERROR!")
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
pin_1 = input("Primer input: ")
pin_2 = input("Segundo input: ")
pin_3 = input("Tercer input: ")


if pin_1 == pin_2 and pin_1 == pin_3 and pin_2 == pin_3:
  print("Son todos iguales")
elif pin_1 == pin_2 or pin_1 == pin_3 or pin_2 == pin_3:
  print("Al menos 2 son iguales")
else:
  print("No hay 2 iguales")
````
