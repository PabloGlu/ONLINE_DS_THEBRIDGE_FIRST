## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
var_list = [4, 6, 8, -1]

var_max = max(var_list)
var_sum = sum(var_list)

var_float = 63.451256965
var_abs = -74



print(f"\n 1. Calcula el máximo de la lista [4, 6, 8, -1]: {var_max}")

print(f"\n 2. Suma todos los elementos de la lista anterior: {var_sum}")

print(f"\n 3. Redondea este float a 3 dígitos decimales 63.451256965: ", round(var_float, 3))

print(f"\n 4. Valor absoluto de -74 ->", abs(var_abs))

```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
var_3 = [7, 10, -2]

print(f"\n 1. Crea una lista con 3 elementos numéricos ->", f"\n {var_3}")

var_3.append(44)

print(f"\n 2. Añade un cuarto elemento ->", f"\n {var_3}")

var_suma = sum(var_3)

print(f"\n 3. Calcula la suma de todos ->", f"\n {var_suma}")

var_del = var_3.pop(2)

print(f"\n 4. Elimina el segundo elemento de la lista ->", f"\n Elemento borrado: {var_del}", f"\n Lista: {var_3}")

var_3.insert(2, -12)

print(f"\n 5. Añade otro elemento en la posicion 3 de la lista ->", f"\n Añadimos -12: {var_3}")

var_0 = [1, 3, 5, 7]

var_3.extend(var_0)

print(f"\n 6. Crea otra lista con 4 elementos y concaténala a la que ya tenías. ->", f"\n {var_0}", f"\n {var_3}")

var_sorted = sorted(var_3) # Un método para ordenar (Crea una lista nueva, no modifica la original)

print(f"\n 7. Ordena la lista de menor a mayor. ->", f"\n {var_sorted}")

var_3.sort(reverse=True) # Otro método para ordenar (Modifica la lista original)

print(f"\n 8. BONUS: Ordena la lista de mayor a menor. ->", f"\n {var_3}")


```

```python
# Experimento, quería ver si podía imprimir la operación completa de la suma, esto es lo mejor que he podido conseguir.

suma_exp = 0
print(f"Printing each number in {var_3}")

for i in var_3:
  print(i, "+")
  suma_exp += i

print(f"Suma: {suma_exp}")
```
