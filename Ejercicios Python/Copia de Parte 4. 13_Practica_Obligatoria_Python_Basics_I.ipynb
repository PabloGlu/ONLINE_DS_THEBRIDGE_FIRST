## Parte II

Crea un programa que realice las siguientes acciones:

1. Solicite al usuario que introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3).
2. Convierta esa entrada en una lista de números enteros.
3. Utilizando funciones built-in, determine el número máximo, el número mínimo y la suma total de los números introducidos.
4. Determine la verdad o falsedad de la siguiente afirmación: "El número máximo menos el número mínimo es igual a la suma total dividida por 5".
5. Muestre los resultados utilizando print.

```python
data_in = input("Introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3): ").replace(" ", "").split(',')
print(f"1. Respuesta \n{data_in}" )

# He intentado algo demasido complejo, voy a intentar algo más sencillo. (ver final)

data_numbers = []

for i in data_in:
  data_numbers.append(int(i))

print(f"2. Lista de números enteros \n {data_numbers}")

var_max = max(data_numbers)
var_min = min(data_numbers)
var_sum = sum(data_numbers)

print(f"3. \n Número máximo: {var_max} \n Número mínimo: {var_min} \n Suma: {var_sum}")

# "El número máximo menos el número mínimo es igual a la suma total dividida por 5".
# Los valores a introducir para comprobrar el true son [1,2,2]

if var_max - var_min == var_sum / 5:
  print(f"\n ¡Es verdad! El número máximo menos el número mínimo es igual a la suma total dividida por 5.")
else:
  print(f"\n ¡No es verdad! El número máximo menos el número mínimo NO es igual a la suma total dividida por 5.")



```
